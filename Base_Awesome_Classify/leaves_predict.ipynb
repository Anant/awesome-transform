```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from collections import Counter
from nltk.tokenize import word_tokenize 
from nltk import NaiveBayesClassifier
leaves_df = pd.read_csv('Archive_articles.csv',delimiter=';', header = 0, error_bad_lines=False)
leaves_df.head()
```

```python
leaves_df.shape
```

```python
sampel_leaves = leaves_df.sample(frac=0.2)
sampel_leaves.head()
```

```python
sampel_leaves.shape
```

```python
t_list = []
tag_list = sampel_leaves['Tags'].tolist()
for tag in tag_list: 
    sp_t = str(tag).split(', ')
    t_list.extend(sp_t)
Counter(t_list)
```

```python
train = list(zip(sampel_leaves.Content, sampel_leaves.Tags))
```

```python
all_words = set(word.lower() for passage in train for word in word_tokenize(str(passage[0])))
t = [({word: (word in word_tokenize(str(x[0]))) for word in all_words}, x[1]) for x in train]
```

```python
classifier = NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```python
print(nltk.classify.accuracy(classifier, t))
```
